### 2. Variables

Variables are used to store data values. Python is dynamically typed, meaning you don't need to declare the type of a variable when you create it. Python automatically assigns the type based on the value.

### 3. Collections and More Data Types

Python offers several built-in data structures to group and organize data. Here are some of the most common ones:

```
# List: An ordered, mutable collection of items. Can contain different data types.
my_list = [1, 2, 'three', True, 4.0]
print("My List:", my_list)
print("Type of my_list:", type(my_list))
print("First element of my_list:", my_list[0]) # Accessing elements by index

# Tuple: An ordered, immutable collection of items. Similar to lists but cannot be changed after creation.
my_tuple = (10, 20, 'thirty', False)
print("\nMy Tuple:", my_tuple)
print("Type of my_tuple:", type(my_tuple))
print("Second element of my_tuple:", my_tuple[1])

# Dictionary: An unordered collection of key-value pairs. Keys must be unique and immutable.
my_dict = {'name': 'Alice', 'age': 30, 'city': 'New York'}
print("\nMy Dictionary:", my_dict)
print("Type of my_dict:", type(my_dict))
print("Name in my_dict:", my_dict['name']) # Accessing values by key

# Set: An unordered collection of unique items. Useful for mathematical set operations.
my_set = {1, 2, 3, 2, 4}
print("\nMy Set:", my_set) # Duplicate '2' is automatically removed
print("Type of my_set:", type(my_set))

# NoneType: Represents the absence of a value or a null value.
my_none = None
print("\nMy None:", my_none)
print("Type of my_none:", type(my_none))
```

```
# Assigning different data types to variables
my_string = "This is a string"
my_integer = 100
my_float = 3.14
my_boolean = False

# Printing the variables and their types
print(my_string)
print(type(my_string))

print(my_integer)
print(type(my_integer))

print(my_float)
print(type(my_float))

print(my_boolean)
print(type(my_boolean))
```
